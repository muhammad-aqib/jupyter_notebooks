```python

garbage=['http','www','this cookies','these cookies','cookies are']

def check_error(s):
    error=[]
    try:
        if s[0].islower():
            error.append('Purpose does not start with a capital letter.')
        if s[-1] != '.':
            error.append('Purpose does not end with a full stop.')
        if '  ' in s:
            error.append('Contains double space in the purpose')
        for i in garbage:
            if i in s.lower():
                error.append('Might contain a url or plural cookie')
                break;
    except:
        print("exception occured for", s)

    return ' '.join(error)
```

```python
error_iterrows=[]

for i,row in csv_remove_unworked.iterrows():
#     print(row['cookie_purpose'],type(row['cookie_purpose']))
#     exit()
    error=check_error(row['cookie_purpose'])
    error_iterrows.append(error)

csv_remove_unworked['error_list'] = error_iterrows
print(csv_remove_unworked.shape)
csv_remove_unworked = csv_remove_unworked[csv_remove_unworked.error_list != '']
csv_remove_unworked.sort_values(by='picked_by')


    
```
