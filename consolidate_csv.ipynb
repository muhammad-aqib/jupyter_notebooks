```python
from usp.tree import sitemap_tree_for_homepage
```

```python
final_consolidated = pd.read_csv('~/Documents/anaconda_work/saas_crawl_consolidated_csv.csv')
len(final_consolidated)
batch = final_consolidated.iloc[0:5]
final_consolidated.columns
```

```python
columns = ['saas_directory','app_name','url','registered_domain','collected_at','sitemap']
values = []
count = 0

for index, row in batch.iterrows():
    url = row['url']
    tree = sitemap_tree_for_homepage(url)

    for page in tree.all_pages():
        values.append([row['saas_directory'],row['app_name'],row['url'],row['registered_domain'],row['collected_at'],page.url])
        count = count + 1
    if count == 0:
        values.append([row['saas_directory'],row['app_name'],row['url'],row['registered_domain'],row['collected_at'],''])
    count = 0

sitemap_df=pd.DataFrame(values,columns=columns)
sitemap_df.to_csv('~/Documents/anaconda_work/final_with_sitemaps.csv', mode='a', header=False)

        
```
